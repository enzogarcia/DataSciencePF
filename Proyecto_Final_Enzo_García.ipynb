# 9 - Primera aproximación:

En esta sección, probaremos con DecissionTreeClasiffier con los datos "en crudo", para ver cómo se comporta el modelo con nuestros datos. Iremos refinando los datos a medida que se avance en el proyecto.

```python
#Asignación de variables

X = dfdiabetes.drop('Diabetes', axis = 1)

Y = dfdiabetes['Diabetes']

#Divido los datos en training y testing

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 10000, stratify = Y)

print(X_train)
```

```python
#Uso DecisionTreeClassifier y entreno el modelo

DTC = DecisionTreeClassifier(random_state = 10000)
DTC.fit(X_train, Y_train)

#Predigo el modelo

Y_pred = DTC.predict(X_test)

#Calculos algunas métricas

accuracy = accuracy_score(Y_test, Y_pred)
matriz_conf = confusion_matrix(Y_test, Y_pred)

# Imprimo los resultados

print(f"Accuracy: {accuracy}")

#Visualizo la matriz de confusión

sns.heatmap(matriz_conf, annot=True, fmt='d', cmap='Blues', cbar=False)
```

Vemos que, con los datos en crudo (solamente se han removidolos valores duplicados), al modelo le cuesta predecir correctamente, sobre todo en diabetes tipo 1 y comete errores groseros prediciendo los demás tipos. De todos modos, aún hay un margen de mejora y se van a utilizar las herramientas vistas en el curso para intentar mejorar las métricas, dado que, al tratarse de una enfermedad, es imperativo ser lo más preciso posible en su diagnóstico.

A continuación, se aplicará escalado de datos en las features tipo float para observar si hay mejoras. Las variables categóricas permanecerán sin alterarse.



```python
#Escalo las variables

scaler = StandardScaler()

numerical_columns = X_train.select_dtypes(include=['float64']).columns

categorical_columns = X_train.select_dtypes(include=['category']).columns

#Escalo solamente las features tipo float

X_train_escaled = scaler.fit_transform( X_train.select_dtypes(include=['float64']))

X_test_escaled = scaler.transform(X_test.select_dtypes(include=['float64']))

#Convierto los valores escalados en train y test (float) a dataframe

X_train_escaled_df_float = pd.DataFrame(X_train_escaled, columns = numerical_columns)

X_test_escaled_df_float = pd.DataFrame(X_test_escaled, columns = numerical_columns)

#Preparo las features categóricas para concatenarlas al dataframe final

X_train_escaled_df_categorical = X_train.select_dtypes(include=['category']).reset_index(drop=True)

X_test_escaled_df_categorical = X_test.select_dtypes(include=['category']).reset_index(drop=True)

#Finalmente concateno los dataframe para training y testing

X_train_final = pd.concat([X_train_escaled_df_float, X_train_escaled_df_categorical], axis=1)

X_test_final = pd.concat([X_test_escaled_df_float, X_test_escaled_df_categorical], axis=1)

print(X_train_final)
```

```python
#Realizo los mismos pasos con los datos escalados

Y_train_escaled = Y_train.copy()
Y_test_escaled = Y_test.copy()

#Entreno el modelo

DTC_escaled = DecisionTreeClassifier(random_state = 10000)
DTC_escaled.fit(X_train_final, Y_train_escaled)

#Predigo el modelo

Y_pred_escaled = DTC_escaled.predict(X_test_final)

#Calculos algunas métricas

accuracy = accuracy_score(Y_test_escaled, Y_pred_escaled)
matriz_conf_escaled = confusion_matrix(Y_test_escaled, Y_pred_escaled)

# Imprimo

print(f"Accuracy: {accuracy}")

#Visualizo la matriz de confusión

sns.heatmap(matriz_conf_escaled, annot=True, fmt='d', cmap='Blues', cbar=False)
```

Se observa que escalar los datos no modifica significativamente el accuracy ni la matriz de confusión, de modo que se pasará a aplicar SMOTE, para ver si los resultados mejoran.

Ahora veremos si mejoran los resultados aplicando oversampling, dado que los datos que representan la diabetes tipo 1 son limitados en relación a los otros grupos. Incluso el tipo 2 necesita una sobrerrepresentación.

Ahora aplicaremos distintas estrategias para el tratamiento de outliers, para ello directamente eliminanermos los outliers, los reemplazaremos por los valores máximos y mínimos de Q1 y Q3, reemplazaremos por la media y aplicaremos Cox-Box, de modo de determinar si alguna de estas estrategias mejora el modelo.

##10 - 1 Estrategia uno: eliminar outliers.

```python
#Verifico y elimino outliers usando Q1 y Q3

columnas_numericas = dfdiabetes_balanced.select_dtypes(include=['float64'])

columnas_categoricas = dfdiabetes_balanced.select_dtypes(include=['category'])

#Creo una máscara booleana que usaremos para almacenar los valores dentro de Q1 y Q3

mask = pd.Series([True] * len(dfdiabetes_balanced), index=dfdiabetes_balanced.index)

# Calculo los percentiles para cada columna float64 usando un bucle

for column in columnas_numericas.columns:
    Q1 = columnas_numericas[column].quantile(0.25)
    Q3 = columnas_numericas[column].quantile(0.75)
    IQR = Q3 - Q1
    limite_inferior = Q1 - 1.5 * IQR
    limite_superior = Q3 + 1.5 * IQR
    mask &= (columnas_numericas[column] >= limite_inferior) & (columnas_numericas[column] <= limite_superior)

#dfdiabetes_sin_outliers = dfdiabetes_sin_outliers.dropna()

#Uso la máscara para crear el dataframe sin outliers

dfdiabetes_sin_outliers = dfdiabetes_balanced[mask]

# Restablezco el índice

dfdiabetes_sin_outliers = dfdiabetes_sin_outliers.reset_index(drop=True)

#Imprimo

print(dfdiabetes_sin_outliers)

```

```python
#Asignación de variables utilizando el dataframe con los outliers eliminados

X = dfdiabetes_sin_outliers.drop('Diabetes', axis = 1)

Y = dfdiabetes_sin_outliers['Diabetes']

#Divido los datos en training y testing

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.3, random_state = 10000, stratify = Y)

print(X_train)
```

```python
#Verifico y trato outliers usando Q1 y Q3, voy a reemplazar los outliers encontrados por el valor máximo o mínimo establecidos por el rango intercuartil

columnas_numericas = dfdiabetes_balanced.select_dtypes(include=['float64'])

columnas_categoricas = dfdiabetes_balanced.select_dtypes(include=['category'])

dfdiabetes_q1yq3 = dfdiabetes_balanced.copy()


# Calculo los percentiles para cada columna float64, luego reemplazo los outliers por Q1 y Q3
for column in columnas_numericas.columns:
    Q1 = columnas_numericas[column].quantile(0.25)
    Q3 = columnas_numericas[column].quantile(0.75)
    IQR = Q3 - Q1
    limite_inferior = Q1 - 1.5 * IQR
    limite_superior = Q3 + 1.5 * IQR
    dfdiabetes_q1yq3[column] = dfdiabetes_q1yq3[column].clip(lower=limite_inferior, upper=limite_superior)

#Imprimo

print(dfdiabetes_q1yq3)
```

##10 - 3 Reemplazo de outliers por el valor medio

```python
#Verifico y trato outliers usando Q1 y Q3, voy a reemplazar los outliers encontrados por el valor medio de cada columna

columnas_numericas = dfdiabetes_balanced.select_dtypes(include=['float64'])

columnas_categoricas = dfdiabetes_balanced.select_dtypes(include=['category'])

dfdiabetes_valor_medio = dfdiabetes_balanced.copy()


# Calculo los percentiles para cada columna float64, luego reemplazo los outliers por los valores medios de cada columna
for column in columnas_numericas.columns:
    Q1 = columnas_numericas[column].quantile(0.25)
    Q3 = columnas_numericas[column].quantile(0.75)
    IQR = Q3 - Q1
    limite_inferior = Q1 - 1.5 * IQR
    limite_superior = Q3 + 1.5 * IQR
    media_columna = columnas_numericas[column].mean()

    # Reemplazar outliers por la media
    dfdiabetes_valor_medio[column] = np.where(
        (columnas_numericas[column] < limite_inferior) | (columnas_numericas[column] > limite_superior),
        media_columna,
        columnas_numericas[column]
    )

#Imprimo

print(dfdiabetes_valor_medio)
```

```python
# Función para asegurarnos el correcto seteo de semilla

def seteo_de_semilla(seed_value):
  # 1. Setear variable de entorno `PYTHONHASHSEED` a un valor fijo
  os.environ['PYTHONHASHSEED']=str(seed_value)

  # 2. Setear semilla propia de generación aleatoria de `python`
  random.seed(seed_value)

  # 3. Setear semilla propia de generación aleatoria de `numpy`
  np.random.seed(seed_value)

  # 4. Setear semilla propia de generación aleatoria de `tensorflow`
  tf.random.set_seed(seed_value)

  # 5. Configurar sesión global de `tensorflow`
  session_conf = tf.compat.v1.ConfigProto(intra_op_parallelism_threads=1, inter_op_parallelism_threads=1)
  sess = tf.compat.v1.Session(graph=tf.compat.v1.get_default_graph(), config=session_conf)
  tf.compat.v1.keras.backend.set_session(sess)
```

```python
#Seteo de semilla

#seteo_de_semilla(10000)


#Encodeo la variable Y

Y_encoded = to_categorical(Y, num_classes = 3)


#Divido en training y testing

X_train_multi, X_test_multi, Y_train_multi, Y_test_multi = train_test_split(X, Y_encoded, test_size = 0.2, random_state = 10000)


#Creo un modelo de tres capas

model = Sequential()
model.add(Dense(10, input_dim = X_train.shape[1], activation = 'relu'))
model.add(Dense(10, activation = 'relu'))
model.add(Dense(3, activation = 'softmax'))

# Compilo el modelo

model.compile(loss = 'categorical_crossentropy', optimizer = 'adam', metrics = ['accuracy'])

# Entreno el modelo
model.fit(X_train_multi, Y_train_multi, epochs=50, validation_split=0.2, verbose=0)

# Evalúo el modelo
loss, accuracy = model.evaluate(X_test_multi, Y_test_multi)

#Imprimo el accuracy

print(f'Loss: {loss}, Accuracy: {accuracy}')
```

#13 - PCA (Principal Components Analysis)

En esta sección, aplicaremos PCA para reducir la dimensionalidad del problema, y analizar qué componentes debemos tomar para tener una varianza adecuada. Una vez realizado esto último, aplicaremos los mismos algoritmos que los aplicados anteriormente (Catboost, XGBoost, LGBM y RandomForest), con Bayes Search para encontrar los mejores hiperparámetros, de modo de observar cómo performan. Descarto el uso de redes neuronales porque no ha sido parametrizada de manera adecuada y, por ello, dio un accuracy bajo.

```python
#Asignación de variables

X = dfdiabetes_balanced.drop('Diabetes', axis = 1)

Y = dfdiabetes_balanced['Diabetes']

#Escalo las variables

scaler = StandardScaler()

X_escalado = scaler.fit_transform(X)

dfescalado = pd.DataFrame(X_escalado, columns = X.columns)

dfescalado = pd.concat([dfescalado, Y.reset_index(drop=True)], axis=1)

#Imprimo el dataframe escalado

print(dfescalado)
```

```python

X = dfescalado.drop('Diabetes', axis = 1)

Y = dfescalado['Diabetes']

#Implemento PCA

pca = PCA()

X_pca = pca.fit_transform(X)

#Debo volcar los datos a un dataframe, dado que se devuelve como un objeto de Numpy

dfpca = pd.DataFrame(X_pca)

#Nombro las columnas

dfpca.columns = ['CP_'+str(i) for i in range(dfpca.shape[1])]

dfpca = pd.concat([dfpca, Y.reset_index(drop=True)], axis = 1)

print(dfpca)
```

```python
# Observo la varianza de cada componente principal

varianza_explicada = pca.explained_variance_ratio_

# Grafico

plt.figure(figsize=(12,8))
plt.bar(range(1, len(varianza_explicada) + 1), varianza_explicada)
plt.xticks(range(1, len(varianza_explicada) + 1))
plt.show()

print(varianza_explicada)
```

```python
#Creo un dataframe con los cinco componentes principales de mayor varianza

vector_pc = ['CP_1', 'CP_2', 'CP_3', 'CP_4', 'CP_5']

dfpca_componentes = dfpca[vector_pc].copy()

#Concateno con mi variable a predecir

dfpca_componentes = pd.concat([dfpca_componentes, Y.reset_index(drop=True)], axis = 1)

print(dfpca_componentes)
```
